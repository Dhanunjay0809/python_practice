### What are attributes 
Attributes are varibales that belong to a class or its objects. 
They store data for the class or objects. 

types of attributes:
1. Instance Attribute   - Unique to each object 
2. Class Attribute  - Shared by all object of the class






### 1. Instance Attribute 
These are variables that are unique to each object. 
Every object has its own copy of instance attributes. 
- Defined inside __init__ method 
- self.attribute_name syntax 
- Each object has diff value 
- Most commonly used type of attribuite

```python
# syntax for instance attribute
class ClassName:
    
    class_attribute = value #class attribute
    
    def __init__(self):
        self.instance_attribute = value # instance attribute
```

```python
# example 

class Student:
    # class attribute = Shared by all studnets
    school_name = "Random Public school"
    total_students = 0          # counter for all studnets 
    
    def __init__(self, name, roll_no, grade):
        # instance attirbues - unique to each student 
        self.name = name 
        self.roll_no = roll_no
        self.grade = grade
        
        # Increment class attribute when new student is created 
        Student.total_students += 1 
        
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Roll no: {self.roll_no}")
        print(f'Grade: {self.grade}')
        print(f"School :{Student.school_name}")    # Accessing class attribute
    
```

### Types of methods 

- in python there are three types of methods 

1. Instance Methods - Work with instance data  
2. Class Methods - Work with class data 
3. Static Methods   - Utility methods(no access to instance or class data)

### Instance Methods 
- These are regular methods that work with instance attributes. 
- they take **self** as that first parameter. 
- Can access and modify instance attributes 
- Can access class attributes 
- Most common type of attribute

```python
# Ex 

class Calculator:
    # construtor
    def __init__(self, brand):
        self.brand = brand 
        self.result = 0 
        
    # Intance methods 
    def add(self, a, b):
        self.result = a + b 
        return self.result 
    
    def display_result(self):
        print(f"{self.brand} Calculator Result: {self.result}")
        
```

```python
calc = Calculator("Casio")
```

```python
calc.add(10, 20)
```

```python
calc.display_result()
```

### Class methods 
they work with class attributes rather tahn instance attributes. They use the @classmethod 
decorator and take **cls** as the first parameter


#### charracteristcs:

- Use @classmethod decorator
- first paramter is cls(represents the class itself)
- Can access/modify class attributes 
- cannot access instance attributes direclty 
- often used as alternative constructors (factory methods)

```python
class Employee:
    company = "Google"
    
    def change_company(self, new_name):
        Employee.company = new_name
```

```python
e1 = Employee()
e2 = Employee()

e1.change_company("Microsoft")

print(Employee.company)  # Microsoft
print(e2.company)        # Microsoft
```

```python
class Manager(Employee):
    pass

Manager().change_company("Amazon")
# Still modifies Employee.company ❌
```

You are directly modifying the class variable using Employee.company

It works because class variables belong to the class, not the instance

- Problem

Instance methods are meant to change instance data

You’re hard-coding the class name (Employee)

Breaks inheritance

```python
# Using @classmethod (✅ correct way)

class Employee:
    company = "Google"
    
    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name
```

```python
e1 = Employee()
e2 = Employee()

Employee.change_company("Microsoft")

print(Employee.company)  # Microsoft
print(e1.company)        # Microsoft
```

Why this is better

cls automatically refers to the calling class

Works correctly with inheritance

Clean OOP design

```python
class Manager(Employee):
    pass

Manager.change_company("Amazon")

print(Manager.company)   # Amazon
print(Employee.company)  # Google
```

```python
# sytax for class method

@classmethod 
def method_name(cls, param1, param2):
    # Access class attributes 
    cls.class_attribute = value 
    return cls(param1, param2)
```

```python
# create 3 classes which coffee espresso, latte, sweet_cappucino

class coffee:
    def __init__(self, name, milk, sugar):
        self.name = name 
        self.milk = milk 
        self.sugar = sugar 
        
    @classmethod 
    def espresso(cls):
        return cls("Espresso", milk = False, sugar = False)
    
    @classmethod 
    def latte(cls):
        return cls("latte", milk = True, sugar = False)
    
    @classmethod 
    def sweet_cappucino(cls):
        return cls("sweet_cappucino", milk = True, sugar = True)
    
    
```

```python
# using factory method 

drink1 = coffee.espresso()
drink2 = coffee.latte()
drink3 = coffee.sweet_cappucino()
```

```python
print(drink1.name, drink1.milk, drink1.sugar)
print(drink2.name, drink2.milk, drink2.sugar)
print(drink3.name, drink3.milk, drink3.sugar)
```

class methods and attribtues 

attributes 
- instance and class attribteu 

methods 
- instance(self), class_method(cls), static(not have cls, self)

```python
class Employee:
    # class attributes 
    # shared by all object of this class
    company_name = "Edukron"
    
    # constructor
    def __init__(self,name, salary):
        # Instance attributes
        # Each object get its own copy
        self.name = name 
        self.salary= salary 
        
        
    # Instance method
    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Salary: {self.salary}")
        print(f"Company: {Employee.company_name}")
        
    # Class method
    # Works with class data(uses cls)
    @classmethod
    def change_company(cls, new_company):
        cls.company_name = new_company
        
    # Static method
    # Utility method
    # Does not use self  or class 
    @staticmethod
    def is_high_salary(salary):
        return salary > 50000
```

```python
emp1 = Employee("Alice", 60000)
emp2 = Employee("Bob", 40000)
```

```python
# using instance method 

emp1.display_details()

```

```python
# usign static method
print(Employee.is_high_salary(emp1.salary))
print(Employee.is_high_salary(emp2.salary))
```
