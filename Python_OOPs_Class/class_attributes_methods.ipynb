```python
# Object- Oriented Approach: 

class BankAccount:
    # construtor - initializes the object 
    
    def __init__(self, account_number, account_holder, balance):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance
        
    # Methods (functions inside a class)
    def deposit(self, amount):
        self.balance += amount 
        print(f"Deposited: {amount}. New Balance: {self.balance}")
        
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"withdraw: {amount}. New balance; {self.balance}")
        else:
            print("Insuffient funds!")
        
    def check_balance(self):
        print(f"Current Balance: {self.balance}")
```

```python
# Creating objects (instances)

account1 = BankAccount("ACC001", "LOHIT KUMAR", 5000)
account2 = BankAccount("ACC002","MOHIT KUMAR", 3000 )
```

## INSTANCE ATTRIBUTE   

Instance attributes are varibale that belong to a specifinc instance (object) of a class. 
Each object has its own copy of instnacne attribute. 

key points: 

- Defined inside __init__() method using self.attribute_name
- Each instance store its own unique values
- Diff objs can have diff values for same attribute
- Modified using self.attribute_name


## Sytax:

def __init__(self, param1, param2):
    self.attribute1 = param1        # Instance attribute
    self.attribute2 = param2        # Instance attribute

```python
class Student:
    
    bank_name = "Kotak Bank"
    
    def __init__(self, name , age, grade):
        self.name = name        # Each student has diff name
        self.age = age          # Each student has diff age
        self.grade = grade     # Each student has diff grade
        
    
    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")
        
    def have_birthday(self):
        self.age += 1
        print(f"{self.name} is not {self.age} years old!")
```

## CLASS ATTRIBUTES

Class attributes are variables that belong to the class itself, not to any specific instance. All instance will share the same copy of class attributes. 

- Defined directly in the class body (outside any method)
- Shared by ALL instances of the class 
- Only ONE copy exitst in memory 
- Modified using ClassName.attribute_name


```python
# SYNTAX 

class MyClas:
    class_attribute = Value        # class attribute
    
    def __init__(self):
        self.instance_attribute = value          # Instnace attribute 
        
        
```

```python
class Employee:
    
    # class attributes - Shared by ALL employee
    company_name = "TechCorp"       # Same for all employees
    employee_count = 0              # Track total employees
    bonus_percentage = 10           # Common bonus rate
    office_location = "Bangalore"   # share location 
    
    def __init__(self, name, salary):
        # Instance attributes - unique to each employee
        self.name = name
        self.salary = salary
        # Modify class attribute when new employee created
        Employee.employee_count += 1
        self.employee_id = Employee.employee_count
        
        
    def display_info(self):
        """Display both instance and class attributes"""
        print(f"\nEmployee ID: {self.employee_id}")
        print(f"Name: {self.name} (instance attribute)")
        print(f"Company: {Employee.company_name} (class attribute)")
        print(f"Office: {Employee.office_location} (class attribute)")
        print(f"Salary: ${self.salary}")
        print(f"Bonus: ${self.salary * Employee.bonus_percentage / 100}")
    
```

```python
emp1 = Employee("John", 50000)
emp2 = Employee("Sarah", 60000)
emp3 = Employee("Mike", 55000)
```

```python
print(f"Total Employees: {Employee.employee_count}")
print(f"Company Name: {Employee.company_name}")

emp1.display_info()
```

## INSTANCE METHOD 

Instance methods are functions defined inside a class that operate on instance of that class. 
They can access and modify instance attributes. 

- Most common types of method in OOP
- First parameter is alwasy "self" (refere to the instance)
- Can access and modify instance attributes using self.attribute
- Can access class attributes using ClassName.attribute or self.attribute 
- Call using object_name.method_name()



```python
# Sytnax 
    def method_name(self, param1, param2):
        # Access instance attribute 
        self.attribute = value 
        
        
```

## CLASS METHODS 

- Class methods are methods that are bound to the class. not to instances. 
- They can access and modify class attributes but cannot access intance attributes. 

keypoints; 
- created using decorater with @classmethod
- First parameter is 'cls' (refers to the class itslef, not instance)
- can access and modify class attributes 
- Cannot access istance attributes (no self availabel)
- Calling using ClassName.method_name() 
- Often used to modify class-level data 

```python
## SYNTAX ; 

@classmethod
    def method_name(cls, param1, param2):
        #Access call attributes 
        cls.class_attribute = Value 
        # Retrun instance if factory method 
        return cls(param1, param2)
    
    
    
```

```python
class Coffee:
    def __init__(self, name, milk, sugar):
        self.name = name
        self.milk = milk 
        self.sugar = sugar
        
    # factory method for espresso 
    @classmethod
    def espresso(cls):
        return cls("Espresso", milk = False, sugar = False)
    
    # factory method for latte
    @classmethod
    def latte(cls):
        return cls("Latte", milk = True, sugar = False)
    
    # factory method for sweet_cappuccino
    @classmethod
    def sweet_cappuccino(cls):
        return cls("Cappuccino", milk = True, sugar = True)
    
```

```python
## Using factory methods 
drink1 = Coffee.espresso()
drink2 = Coffee.latte()
drink3 = Coffee.sweet_cappuccino()
```

```python
print(drink1.name, drink1.milk, drink1.sugar)
print(drink2.name, drink2.milk, drink2.sugar)
print(drink3.name, drink3.milk, drink3.sugar)
```

```python
class Product: 
    # Class attribute
    tax_rate = 0.1 
    total_products = 0 
    product_list = []
    
    
    def __init__(self, name, price):
        # Instance attribute
        self.name = name
        self.price = price 
        # Update Class attributes 
        Product.total_products += 1 
        Product.product_list.append(name)
        
    # Class method - modifies class attributes 
    @classmethod 
    def update_tax_rate(cls, new_rate):
        old_rate = cls.tax_rate
        cls.tax_rate = new_rate
        print(f"Tax rate update {old_rate*100}% to {new_rate*100}%")
        print(f"This affects all {cls.total_products} products!")
        
    # class method 
    @classmethod
    def from_string(cls, product_string):
        """
        Creats product from string format: "name-price" 
        This is a FACTORY METHOD- alternative way to creat oject 
        Returns: new Product instance
        """
        
        name, price = product_string.split("-")
        print(f'Creating product from string: {product_string}')
        return cls(name, float(price))      
    
    # Another factory method 
    @classmethod
    def from_dict(cls, product_dict):
        """Creates product from dicotry"""
        return cls(product_dict["name"], product_dict['price'])
    
    # Class method - accsing class data 
    @classmethod 
    def get_total_products(cls):
        return cls.total_products
    
    @classmethod
    def display_all_products(cls):
        print(f"\nTotal products: {cls.total_products}")
        
        
    # Instance method using class attribute
    def calculate_final_price(self):
        """Calculates price including tax"""
        return self.price * (1 + Product.tax_rate)
    
    def display(self):
        """Shows product information"""
        print(f"{self.name}: ${self.price:.2f} "
              f"(Final: ${self.calculate_final_price():.2f})")


        
        
# /Users/mac/Desktop/Python class/Sept_Python_online/python_class/class_attributes_methods.ipynb
```
