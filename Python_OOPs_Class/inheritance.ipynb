## what is inhertance? 
- It is a mechanism where a new class(child/ derived class/ sub class) can inherit properites and methods from an existing class(parent/ base class). 
- This will promote code reusability and relatiship between class. 

## Why to use Inhertance? 

- 1. Code Reuseablitily. 
- 2. Extensibility.
- 3. Maintainability.


```python
# Basic inhertance syntax 

class ParentClass:
    def __init__(self):
        # parent attributes 
        pass
    
    def parent_method(self):
        # Parent methods
        pass

# Child/Derived class 

class ChildClass(ParentClass):
    def __init__(self):
        super().__init__()  # call parent constructor
        # child-specific attributes 
        pass 
    
    def child_methods(self):
        # child-specific methods 
        pass 
    
    
```

## the super() fucntion 

super() is a built_in fucniton that returns a temporary object of the parent class, allowing you to call its methods


```python
# Animal (eat, sleep, make_sound)

# dog(Animal)       -> (eat, sleep, make_sound)
# cat(Animal)       -> (eat, sleep, make_sound)

# Parent class
class Animal:
    def __init__(self, name, age):
        self.name = name 
        self.age = age
        
    def eat(self):
        print(f"{self.name} is eating...")
        
    def sleep(self):
        print(f"{self.name} is sleeping...")
        
    def make_sound(self):
        print(f"{self.name} is making sound...")
        
```

```python
# Child class 1 

class Dog(Animal):  # dog inherits from Animal
    def __init__(self,name, age, breed):
        super().__init__(name, age)     # Call parent constructor
        self.breed = breed              # Additional attribute 
        
    # New method methods specific to dog 
    def fetch(self):
        print(f"{self.name} is fetching the ball!")
        
    # Override parent method
    def make_sound(self):
        print(f"{self.name} Barks...")
    
        
# child class 2 

class Cat(Animal):  # Cat inherits from Animal
    def __init__(self,name, age, color):
        super().__init__(name, age)     # Call parent constructor
        self.color = color              # Additional attribute  
        
    # New method methods specific to cat
    def scratch(self):
        print(f"{self.name} is scracting the furniture")
        
    # Override parent method
    def make_sound(self):
        print(f"{self.name} meows...")
        
    
```

```python
# Testing 

dog = Dog("Buddy", 3, "Golden Retriever")
cat = Cat("Kitty", 2, "Orange" )
```

```python
# print dog actions 

dog.eat()       # Inherited from Animal
dog.sleep()     # Inherited from Animal
dog.make_sound()    # Overidden in Dog
dog.fetch()         # Specific to dog
```

```python
cat.eat()       # Inherited from Animal
cat.sleep()     # Inherited from Animal
cat.make_sound() # Overidden in Cat
cat.scratch()   # Specific to cat
```

## Single Inheritance

```python
## Parent class

class Vehicle:
    def __init__(self, brand, model):
        print("step 1: Vehicle __init__ called")
        self.brand = brand
        self.model = model 
        print(f"step 2: Set Brand= {brand}, modle = {model}")
        
        
    def display(self):
        # Diplays attribues initialized in vehicle 
        print(f"Vehicle: {self.brand} {self.model}")
```

```python
# child class 

class Car(Vehicle):
    def __init__(self, brand, model, doors):
        print("step 3: Car __init__ called")
        super().__init__(brand, model)
    
        print("step 4: Back in Car __init__")
        
        self.doors = doors
        print(f"Step 5: Set doors= {doors}")
        
        
    def display(self):
        # This method OVERRIDES Vehicle.display()
        print("Car display method:")
        
        # Class Vehicle.display() to reuse parent logic
        super().display()
        
        # Add child-specific behavior 
        print(f"Doors: {self.doors}")        
```

```python
print("Creating car object")

car = Car("Toyota", "Camry", 4)
```

## 1. Multilevel Inheritance

#### Definition

Multilevel inheritance is when a derive class inherits from another derived class, creating a chain or hierarchy of inheritance:

**GrandParent -> Parent -> child -> GrandChild ->.....**



```python
# Multilevel Inheritance 

class GrandParent:
    pass 

class Parent(GrandParent):
    pass 

class Child(Parent):        # Multilevel Inheritance 
    pass
```

```python
class Vehicle:
    """
    Base class representing any vehicle.
    This is the top-level class in our hierarchy.
    """

    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        print(f"Vehicle created: {brand} {model}")

    def start(self):
        """Method to start the vehicle"""
        return f"{self.brand} {self.model} is starting..."

    def stop(self):
        """Method to stop the vehicle"""
        return f"{self.brand} {self.model} is stopping..."
```

```python
class Car(Vehicle):
    """
    Car inherits from Vehicle (Level 2)
    Adds car-specific features like doors
    """

    def __init__(self, brand, model, doors):
        # Call parent class constructor
        super().__init__(brand, model)
        self.doors = doors
        print(f"Car initialized with {doors} doors")

    def honk(self):
        """Method specific to cars"""
        return "Beep! Beep!"

    def open_trunk(self):
        """Cars have trunks"""
        return f"Opening trunk of {self.brand} {self.model}"
```

```python

class ElectricCar(Car):
    """
    ElectricCar inherits from Car which inherits from Vehicle (Level 3)
    This demonstrates multilevel inheritance
    """

    def __init__(self, brand, model, doors, battery_capacity):
        # Call parent class (Car) constructor
        super().__init__(brand, model, doors)
        self.battery_capacity = battery_capacity
        self.charge_level = 100  # fully charged
        print(f"Electric car with {battery_capacity} kWh battery created")

    def charge(self):
        """Method specific to electric cars"""
        self.charge_level = 100
        return f"Charging {self.brand} {self.model}... Battery: {self.battery_capacity} kWh"

    def get_range(self):
        """Calculate estimated driving range"""
        # Simple assumption: 4 miles per kWh
        range_miles = self.battery_capacity * 4 * (self.charge_level / 100)
        return f"Estimated range: {range_miles:.0f} miles"

    # Overriding start method from Vehicle
    def start(self):
        """Electric cars start silently"""
        if self.charge_level > 0:
            return f"{self.brand} {self.model} is starting silently... (Battery: {self.charge_level}%)"
        return "Cannot start - battery is dead!"
```

```python
print("Multilevel inheritance")

tesla = ElectricCar("Tesla", "Model 3", 4, 75)
```

```python
print(tesla.start())
```

### Multiple Inheritance

This is when classs inherits from more than one parent class at the same time.

The child class gains access to attributes and methods from all its parents classes. 

key chars: 

- A class can have multiple direct parent class.
- Inherite features from all parent classes. 
- can lead to the diamond problem
 

```python
# Multiple inhertance exmaple 

class Parent1:   
    pass 

class Parent2:   
    pass 

class Child(Parent1, Parent2):      # Multiple Inhertance
    pass
```

```python
class Employee:     # parent 1
    def __init__(self, emp_id):
        self.emp_id = emp_id 
        
    def work(self):
        return "Employee is working"
    
class Teacher:      # Paernt 2
    def __init__(self, subject):
        self.subject = subject 
    
    def teach(self):
        return f"Teaching {self.subject}"
    
class Professor(Employee, Teacher):
    # child class inheriting from both parent class
    def __init__(self, emp_id, subject):
        # Explicitly calling both parent constructors 
        # Because super() alone cannot automatically initalize both 
        Employee.__init__(self, emp_id)
        Teacher.__init__(self, subject)
    
    def profile(self):
        return f"Professor ID : {self.emp_id}, subject = {self.subject}"
    
    
    

    
```

```python
prof = Professor(101, "Data Science")
```

```python
# Accesing method from Employee parent class
print(prof.work())

# Accesing method from Teacher parent class
print(prof.teach())

# Accesing method from  Professor class itself
print(prof.profile())
```
