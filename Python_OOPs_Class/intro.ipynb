```python
# Object-Oriented Approach:


class BankAccount: 
    def __init__(self, account_number, account_holder, balance):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance
        
    def deposit(self,amount):       # methods are nothing functions inside class
        self.balance += amount 
        print(f"Deposited: {amount}. New balance: {self.balance}")
    
    def withdraw(self,amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawn: {amount}.New balance: {self.balance}")
        else:
            print("Insufficient Funds!")
        
    def check_balance(self):
        print(f"Current balance: {self.balance}")
        
    
```

### The __init__ Constructor. 

- Special method(called a "dunder" method - double underscore)

- Used to initialize the object's attributes with starting values
- Think of it as a 'setup" method

```python
# Define the class 
# i want Car class to have 3 fucntions
# 1. It should display_info about car 
# 2. Accelerate 
# 3. Break
class Car:
    # Constructor - runs automatically when object is created 
    def __init__(self, brand, model, color):
        # Instance attributes - unique to each object 
        self.brand = brand
        self.model = model
        self.color = color
        self.speed = 0      # Default value
        
    # Instance methods
    # This method is for displaying the car details
    def display_info(self):
        print(f"Car: {self.brand}{self.model}")
        print(f"Color: {self.color}")
        print(f"Current Speed: {self.speed} km/h")
        
    # This method is for accelerating   
    def accelerate(self,increase):
        self.speed += increase
        print(f"{self.brand} is now going at {self.speed} km/h")
        
        
    # Thid method is for applying break 
    def brake(self, decrease):
        self.speed -= decrease 
        if self.speed < 0:
            self.speed = 0 
        print(f"{self.brand} slowed down to {self.speed} km/h")
        
        
# Now the class (blueprint for any car) is ready 
```

```python
# for creating class object Car(brand, model, color)


car1 = Car("Toyota", "Camry", "Blue")
car2 = Car("Honda", "Civic", "Red")
car3 = Car("Tesla", "Model 3", "White")

# car1, car2, car3 are the objects of class Car 
# Class = Plan/blueprint 
# Object = Actual item made from the plan 
```
