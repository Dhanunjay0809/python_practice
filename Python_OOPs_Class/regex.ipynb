## Regular Expression. 
we can perform these operations on given string

- Search 
- match 
- extract
- replace 
- validate

The re module in Python is used for working with Regular Expressions (Regex).

Regex is a pattern-matching language that helps you search, extract, validate, replace, and split text efficiently.

Normal string methods → simple text

Regex (re) → powerful text rules

Why do we need re?

Imagine you are working with:

phone numbers

emails

usernames

logs

scraped website text

messy real-world data

String methods (find, split, replace) break down quickly when patterns become complex.

Regex solves that.

```python
# Importing the module
import re
```

Core idea of Regex

A regex pattern describes what kind of text you want, not the exact text.

Examples:

\d → any digit (0–9)

[a-z] → any lowercase letter

\w+ → word characters (letters + digits + _)

^ → start of string

$ → end of string

# Most Important re Functions (90% usage)

### 1. re.search() → Find first match

```python
text = "My phone number is 9876543210"
match = re.search(r"\d{10}", text)

print(match)
```

## 2. re.findall() → Find all matches

```python
text = "Prices are 100, 200, and 300"
numbers = re.findall(r"\d+", text)

print(numbers)

# use when You want all occurrences
```

```python
# 3. re.match() → Match from start only

text = "Python is awesome"
result = re.match(r"Python", text)

print(result)
print(result.group())
```

```python
# 4. re.sub() → Replace text

text = "My number is 9876543210"
masked = re.sub(r"\d", "*", text)

print(masked)


# Use when:
# ✔️ Masking data
# ✔️ Cleaning text
```

```python
# 5. re.split() → Split using pattern

text = "apple,banana;orange mango"
items = re.split(r"[ ,;]", text)

print(items)

# Use when:
# ✔️ Separators are inconsistent
```

## 1. Characters

| Pattern | Meaning |
|--------|---------|
| `a` | matches `'a'` |
| `abc` | matches `'abc'` |
| `.` | any character (except newline) |

---

## 2. Character Classes

| Pattern | Meaning |
|--------|---------|
| `\d` | digit (0–9) |
| `\w` | letter, digit, `_` |
| `\s` | whitespace (space, tab, newline) |
| `\D` | not a digit |
| `\W` | not a word character |
| `\S` | not whitespace |

---

## 3. Quantifiers (How many?)

| Pattern | Meaning |
|--------|---------|
| `*` | 0 or more |
| `+` | 1 or more |
| `?` | 0 or 1 |
| `{3}` | exactly 3 |
| `{2,5}` | between 2 and 5 |

---

## 4. Anchors (Position)

| Pattern | Meaning |
|--------|---------|
| `^` | start of string |
| `$` | end of string |
